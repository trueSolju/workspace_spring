##### Index

```python
import pandas as pd
```

```python
data = {
    '이름':['김자바','이자바','박자바','최캐드','정캐드'],
    '학급':['자바반','자바반','자바반','캐드반','캐드반'],
    '국어점수':[90,80,70,100,90],
    '영어점수':[70,80,80,70,90],
    '평균':[80.0,80.0,75.0,85.0,90.0],
    '취미':['독서','','','요가','축구']
}
```

```python
df=pd.DataFrame(data, index=['1번','2번','3번','4번','5번'])
df
```

```python
# DataFrame에서 index값만 가져오기

df.index

# index에 이름을 부여
df.index.name ='학번'
df
```

```python
#변경한 index를 초기화
# 아래와 같은 코드를 작성하면 index가 컬럼으로 변경 됨.
df.reset_index()
```

```python
# 변경된 index를 완전히 삭제하는 방법
# 매개변수로 drop만 사용하면 drop후의 모습을 보여줄 뿐 실제 drop되진 않음
df.reset_index(drop=True)
df.reset_index(drop=True, inplace=True)
df
```

```python
# 기존 컬럼을 index로 지정
df.set_index('이름',inplace=True)
df
```

```python
# 인덱스를 기준으로 정렬
df.sort_index(ascending=False, inplace=True)
df
```
